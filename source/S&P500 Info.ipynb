# S&P 500 Company List

```python
# Import dependencies
from splinter import Browser
from bs4 import BeautifulSoup
import pandas as pd

# Launch the browser
browser = Browser('chrome')
```

```python
# Set Wikipedia URL
url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"

# Go to the URL
browser.visit(url)

# Save the HTML from the browser
html = browser.html

# Create a BeautifulSoup object
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Extract all table rows
table = soup.find("table", class_="wikitable sortable jquery-tablesorter")
table = table.find("tbody")
# table_rows = table.find_all("td")
table
```

```python
# Empty list to store info
SP500_info = []
```

```python
# Use for loop to extract data
rows = table.find_all("tr")

for row in rows:
    row_data = row.find('td').text.strip()
    SP500_info.append(row_data)
SP500_info
```

```python
# Create a Pandas DataFrame by using the list of rows and a list of the column names
SP500_df = pd.DataFrame(SP500_info)
SP500_df.columns = ['Symbol', 'Security', 'GICS_Sector', 'GICS Sub_Industry', 'HQ Location', 
                    'Date_Added', 'CIK', 'Founded']
SP500_df.head()
```
